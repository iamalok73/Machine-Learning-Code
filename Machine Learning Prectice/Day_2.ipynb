```python
plt.plot(df["year"], df["per capita income (US$)"])
plt.title("The Graph Bt year vs per capita income")
plt.xlabel("year")
plt.ylabel("per capita income")
```

```python
X = df[["year"]]
y = df["per capita income (US$)"] 
```

```python
model = LinearRegression()
```

```python
model
```

```python
model.fit(X,y)
```

```python
slop = model.coef_
print(slop)
```

```python
intercept = model.intercept_
print(intercept)
```

```python
# linear equation 

# y = mx+c

```

```python
# y = slop*year + intercept
```

```python
year = int(input("enter the year : "))

income = slop*year+intercept



print(f"year : {year}\nIncome is :{income} ")
```

```python
y_pred = model.predict(X)
```
